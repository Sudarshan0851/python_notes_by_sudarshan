**Handling the missing data**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([[1,2,3],[4,np.nan,5],[np.nan,np.nan,np.nan]])
df
```

```python
#How we are going to drop the null values
#df.dropna()
df.dropna()
#The above command removes the rows having null values
#remove only the rows having all null values  --> use "how"
```

```python
df.dropna(how="all")
```

```python
#replace the first index column data with null value
df[1] = np.nan
```

```python
#remove the columns which is having alll the null values
df.dropna(axis = 0,how="all")
```

```python
#thresh :- count of non-null values
df = pd.DataFrame([[1,2,3],[4,np.nan,5],[6,np.nan,np.nan]])
#df.dropna(thresh=2) # in a row count of non null values is 2
```

```python
df.dropna(thresh=3,inplace=True)
```

```python
df
```

```python
#there is a dataset of 100 rows and 10 columns
#where in the particular row if the null values is >=4 then delete that row
df.dropna(thresh=6)
```

```python
##there is a dataset of 100 rows and 10 columns
#if the particular column having 20 null values drop that column
df.dropna(thresh=80,axis=1)
```

```python
#df.dropna(axis=0/1,how="all/any",thresh=count,inplace=True/False)
```

**Fill null values**

```python
df = pd.DataFrame([[1,2,3],[4,np.nan,5],[np.nan,np.nan,np.nan]])
df
```

```python
#df.fillna(value)
df.fillna(0)
```

```python
df.fillna(df.mean())
```

```python
#fill the null values with front values
df.fillna(method="ffill")
```

**Data Manipulation in Pandas**

```python
import pandas as pd
import numpy as np

```

```python
df = pd.DataFrame(columns= ['Name','Age','City'])
df
```

```python
#insert the data into dataframe
#1. alix,25,delhi
#to insert the data into dataframe use the "loc"
df.loc[0] = ['alix',25,"Delhi"]
df.loc[1] = ['Bunny',26,"Mumbai"]
```

```python
print(df)
```

```python
#method 2: pd.concat()
new_row = pd.DataFrame([{"Name":"Charlie","Age":25,"City":"Hyderabad"}])
df = pd.concat([df,new_row],ignore_index=True)
df
```
