Apply condition in array's

Np.where(logic,true,false)

```python
a = np.array([[1,2],[3,4],[5,6]])
#if the number more then 2 print those numbers others wise print -1
print(np.where(a>2,a,-1))
```

```python
b = np.array([10,20,21,16,13,17,19])
#apply the condition and check the evenumbers in above data
#if even print even numbers otherwise print -1
print(np.where(b%2==0,b,-1))
```

```python
#from above problem compute the indexes where the condition is true
#np.argwhere(logic)
print(np.argwhere(b%2==0).flatten())  #flatten is used to print output in 1D format
```

```python
#x1 + x2 = 1200
#2x1 + 3x2 = 6000
#find the values of x1 and x2
a = np.array([[1,1],[2,3]])
b = np.array([1200,6000])
x = np.linalg.solve(a, b)
print(x)
```
